### QUESTIONS
### 1. Which AI roles are most in demand?

```python
top_roles = df['job_title'].value_counts().head(10)
print(top_roles)
```

```python
import matplotlib.pyplot as plt

df['job_title'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Most In-Demand AI Roles")
plt.xlabel("Job Role")
plt.ylabel("Number of Jobs")
plt.show()
```

### 2. What skills should job seekers learn?

```python
skills = df[['skills_python','skills_sql','skills_ml','skills_cloud']].sum()
print(skills)

skills.plot(kind='bar')
plt.title("Most Demanded AI Skills")
plt.xlabel("Skills")
plt.ylabel("Number of Jobs Requiring Skill")
plt.show()
```

### 3. How does experience affect salary?

```python
salary_experience = df.groupby('experience_level')['salary'].mean()
print(salary_experience)
salary_experience.plot(kind='bar')
plt.title("Average Salary by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary")
plt.show()
```

### 4. Are AI jobs mostly remote, hybrid, or onsite?

```python
import pandas as pd

df = pd.read_csv("Job_market_data.csv")

# Count jobs by work type
remote_distribution = df['remote_type'].value_counts()

print(remote_distribution)

# Percentage distribution
remote_percentage = df['remote_type'].value_counts(normalize=True) * 100

print(remote_percentage)
```

```python
import matplotlib.pyplot as plt

df['remote_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Distribution of AI Job Work Types")
plt.ylabel("")
plt.show()
```

### Industry Hiring Trends

```python
industry_hiring = df['company_industry'].value_counts()

industry_hiring.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title("AI Hiring by Industry")
plt.xlabel("Industry")
plt.ylabel("Number of Job Postings")
plt.xticks(rotation=45)
plt.show()
```
